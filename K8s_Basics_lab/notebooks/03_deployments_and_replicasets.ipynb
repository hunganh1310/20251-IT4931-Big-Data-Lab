## 1. Deployments

**Deployment** quáº£n lÃ½ ReplicaSets vÃ  Pods:
- Declarative updates cho Pods vÃ  ReplicaSets
- Rolling updates vÃ  rollbacks
- Scaling
- Self-healing (restart failed pods)


```python
import subprocess

def run_kubectl(command):
    try:
        result = subprocess.run(command.split(), capture_output=True, text=True, check=True)
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error: {e.stderr}"

# Deployment manifest
deployment = """
apiVersion: apps/v1
kind: Deployment
metadata:
  name: simple-deployment
spec:
  replicas: 3
  selector:
    matchLabels:
      app: simple-app
  template:
    metadata:
      labels:
        app: simple-app
    spec:
      containers:
      - name: app
        image: python:3.10-slim
        command: ["python", "-m", "http.server", "8000"]
        ports:
        - containerPort: 8000
"""

print("ğŸ“ Deployment Manifest:")
print(deployment)
print("\nğŸ’¡ To create: kubectl apply -f manifests/02-deployment-example.yaml")
```

```python
# Check deployments
print("ğŸ“Š Deployments:")
print(run_kubectl("kubectl get deployments"))

print("\nğŸ“‹ ReplicaSets:")
print(run_kubectl("kubectl get replicasets"))

print("\nğŸ“¦ Pods:")
print(run_kubectl("kubectl get pods"))
```

## 2. Scaling

Scale deployment báº±ng cÃ¡ch thay Ä‘á»•i sá»‘ replicas:
