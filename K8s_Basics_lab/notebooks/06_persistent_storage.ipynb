## 1. Volumes

**Volume** cho ph√©p Pods persist data:
- Data survives pod restarts
- Share data gi·ªØa containers trong c√πng pod
- Types: emptyDir, hostPath, persistentVolumeClaim, etc.


```python
import subprocess

def run_kubectl(command):
    try:
        result = subprocess.run(command.split(), capture_output=True, text=True, check=True)
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error: {e.stderr}"

# Pod v·ªõi volume
pod_with_volume = """
apiVersion: v1
kind: Pod
metadata:
  name: pod-with-volume
spec:
  containers:
  - name: app
    image: python:3.10-slim
    volumeMounts:
    - name: data
      mountPath: /data
  volumes:
  - name: data
    emptyDir: {}
"""

print("üìù Pod v·ªõi Volume:")
print(pod_with_volume)
```

## 2. PersistentVolumes v√† PersistentVolumeClaims

- **PersistentVolume (PV)**: Cluster resource (storage)
- **PersistentVolumeClaim (PVC)**: Request cho storage
- **StorageClass**: Dynamic provisioning


```python
# PVC manifest
pvc = """
apiVersion: v1
kind: PersistentVolumeClaim
metadata:
  name: app-pvc
spec:
  accessModes:
    - ReadWriteOnce
  resources:
    requests:
      storage: 1Gi
"""

print("üìù PersistentVolumeClaim:")
print(pvc)

print("\nüìä Check PVs v√† PVCs:")
print("kubectl get pv")
print("kubectl get pvc")
```
