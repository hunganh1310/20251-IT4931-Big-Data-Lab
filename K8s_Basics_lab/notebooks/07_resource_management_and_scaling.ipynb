## 1. Resource Requests v√† Limits

**Requests**: Guaranteed resources
**Limits**: Maximum resources

- CPU: Measured in cores (m = millicores)
- Memory: Measured in bytes (Mi, Gi)


```python
import subprocess

def run_kubectl(command):
    try:
        result = subprocess.run(command.split(), capture_output=True, text=True, check=True)
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error: {e.stderr}"

# Deployment v·ªõi resources
deployment_with_resources = """
apiVersion: apps/v1
kind: Deployment
metadata:
  name: app-with-resources
spec:
  replicas: 2
  template:
    spec:
      containers:
      - name: app
        image: python:3.10-slim
        resources:
          requests:
            memory: "128Mi"
            cpu: "100m"
          limits:
            memory: "256Mi"
            cpu: "500m"
"""

print("üìù Deployment v·ªõi Resources:")
print(deployment_with_resources)
```

## 2. Horizontal Pod Autoscaler (HPA)

HPA t·ª± ƒë·ªông scale pods d·ª±a tr√™n metrics:
- CPU utilization
- Memory utilization
- Custom metrics


```python
# Create HPA
print("üìà Create HPA:")
print("kubectl autoscale deployment app-with-resources --cpu-percent=70 --min=2 --max=10")

print("\nüìä Check HPA:")
print("kubectl get hpa")

print("\nüìã HPA Details:")
print("kubectl describe hpa app-with-resources")
```

## 3. Resource Quotas

Resource Quotas gi·ªõi h·∫°n resources trong namespace:
- CPU v√† Memory limits
- Number of objects
- Storage limits


```python
# Resource Quota
quota = """
apiVersion: v1
kind: ResourceQuota
metadata:
  name: compute-quota
spec:
  hard:
    requests.cpu: "2"
    requests.memory: 4Gi
    limits.cpu: "4"
    limits.memory: 8Gi
    pods: "10"
"""

print("üìù Resource Quota:")
print(quota)

print("\nüìä Check Quotas:")
print("kubectl get quota")
```
