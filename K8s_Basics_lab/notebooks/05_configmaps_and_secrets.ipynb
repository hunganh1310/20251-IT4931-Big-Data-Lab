## 1. ConfigMaps

**ConfigMap** l∆∞u configuration data:
- Key-value pairs
- Kh√¥ng ch·ª©a sensitive data
- C√≥ th·ªÉ mount v√†o Pods ho·∫∑c d√πng nh∆∞ env vars


```python
import subprocess

def run_kubectl(command):
    try:
        result = subprocess.run(command.split(), capture_output=True, text=True, check=True)
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error: {e.stderr}"

# Create ConfigMap
print("üìù Create ConfigMap:")
print("kubectl create configmap app-config --from-literal=database_host=postgres --from-literal=api_url=https://api.example.com")

print("\nüìã View ConfigMap:")
print("kubectl get configmap app-config -o yaml")
```

## 2. Secrets

**Secret** l∆∞u sensitive data:
- Base64 encoded (kh√¥ng ph·∫£i encrypted!)
- Passwords, API keys, certificates
- T∆∞∆°ng t·ª± ConfigMap nh∆∞ng cho sensitive data


```python
# Create Secret
print("üîê Create Secret:")
print("kubectl create secret generic app-secret --from-literal=api_key=secret123 --from-literal=db_password=pass123")

print("\nüìã View Secret:")
print("kubectl get secret app-secret -o yaml")

print("\nüí° Decode Secret:")
print("kubectl get secret app-secret -o jsonpath='{.data.api_key}' | base64 -d")
```

## 3. Using ConfigMaps v√† Secrets trong Pods

C√≥ th·ªÉ s·ª≠ d·ª•ng ConfigMaps v√† Secrets nh∆∞:
- Environment variables
- Volume mounts
- Command arguments


```python
# Pod v·ªõi ConfigMap v√† Secret
pod_with_config = """
apiVersion: v1
kind: Pod
metadata:
  name: app-with-config
spec:
  containers:
  - name: app
    image: python:3.10-slim
    env:
    - name: DATABASE_HOST
      valueFrom:
        configMapKeyRef:
          name: app-config
          key: database_host
    - name: API_KEY
      valueFrom:
        secretKeyRef:
          name: app-secret
          key: api_key
"""

print("üìù Pod v·ªõi ConfigMap v√† Secret:")
print(pod_with_config)
```
