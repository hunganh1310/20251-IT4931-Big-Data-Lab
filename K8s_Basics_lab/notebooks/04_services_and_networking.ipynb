## 1. Services

**Service** expose Pods ƒë·∫øn network:
- Stable IP address v√† DNS name
- Load balancing
- Service discovery
- Abstract Pods (Pods c√≥ th·ªÉ thay ƒë·ªïi nh∆∞ng Service stable)


```python
import subprocess

def run_kubectl(command):
    try:
        result = subprocess.run(command.split(), capture_output=True, text=True, check=True)
        return result.stdout
    except subprocess.CalledProcessError as e:
        return f"Error: {e.stderr}"

# Service manifest
service = """
apiVersion: v1
kind: Service
metadata:
  name: simple-service
spec:
  selector:
    app: simple-app
  ports:
  - port: 80
    targetPort: 8000
  type: ClusterIP
"""

print("üìù Service Manifest:")
print(service)
print("\nüí° To create: kubectl apply -f manifests/03-service-example.yaml")
```

## 2. Service Types

### LoadBalancer

```python
# Check services
print("üìä Services:")
print(run_kubectl("kubectl get services"))

print("\nüìã Service Details:")
print(run_kubectl("kubectl describe service simple-service"))

# Test service
print("\nüîç Test service:")
print("kubectl run -it --rm test-pod --image=busybox --restart=Never -- wget -qO- http://simple-service")
```
